```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dems = pd.read_csv('../datasets/dem_candidates.csv')
dems.head()
```

# Data Cleaning

```python
endorsement_cols = dems.columns[pd.Series(dems.columns).str.contains('Endorsed')]
dems[endorsement_cols] = dems[endorsement_cols].fillna(0)
dems[endorsement_cols] = dems[endorsement_cols].replace({'No': 0, 'Yes': 1})
dems['Total Endorsements'] = dems[endorsement_cols].sum(1)
dems['Endorsed'] = (dems['Total Endorsements'] > 0).astype(int)
```
